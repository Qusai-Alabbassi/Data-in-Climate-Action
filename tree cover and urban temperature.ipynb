```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# Load Excel file
file_path = r'/Users/qusaialabbassi/Desktop/pythonprojects/Environmental/urban_heat_tree_cover.xlsx'
df = pd.read_excel(file_path)

# Preview the data
print("Sample Data:")
print(df.head())

# Plotting Tree Cover vs Surface Temperature
sns.lmplot(x='Tree Cover %', y='Surface Temperature Â°C', data=df)
plt.title("Tree Cover vs Surface Temperature")
plt.xlabel("Tree Cover (%)")
plt.ylabel("Surface Temp (Â°C)")
plt.tight_layout()
plt.show()

# Linear regression model
X = df[['Tree Cover %']]
y = df['Surface Temperature Â°C']
model = LinearRegression().fit(X, y)

# Output results
print("\n Linear Regression Results:")
print("Slope (Impact per % Tree Cover):", round(model.coef_[0], 3))
print("Intercept (Base Temp):", round(model.intercept_, 2))
print("RÂ² (Explained Variation):", round(model.score(X, y), 2))

# Predict temperature reduction if tree cover increases by 20%
tree_increase = 20
temp_reduction = round(model.coef_[0] * tree_increase, 2)
print(f"\n Estimated temperature reduction with +{tree_increase}% tree cover: {temp_reduction} Â°C")
```
